```python
'''
Strange values in columns with discrepancies:
CRASH CODE - 00, 33, 32, 31, also needs a lot of standardization
IMPACT CODE - 30
PRIMARY CIRCUMSTANCE CODE  - 43

These need to be handled/removed before converting codes-to-descriptions/descriptions-to-codes
'''
```

```python
pd.set_option('display.max_columns', 39)
crash_df[crash_df['PRIMARY CONTRIBUTING CIRCUMSTANCE CODE'] == 43.0]
```

```python
crash_df[crash_df['MANNER OF IMPACT CODE'] == 30.0]
```

```python
crash_df = crash_df[(crash_df['CRASH CLASSIFICATION CODE'] != '33') & 
           (crash_df['CRASH CLASSIFICATION CODE'] != '32') & 
           (crash_df['CRASH CLASSIFICATION CODE'] != '31') &
           (crash_df['CRASH CLASSIFICATION CODE'] != '00')]
crash_df['CRASH CLASSIFICATION CODE'] = '0' + crash_df['CRASH CLASSIFICATION CODE'].str[-1]

crash_df = crash_df[crash_df['MANNER OF IMPACT CODE'] != 30.0]

crash_df = crash_df[crash_df['PRIMARY CONTRIBUTING CIRCUMSTANCE CODE'] != 43.0]

crash_df.head()
```

```python
crash_df['CRASH CLASSIFICATION CODE'].value_counts()
```

```python
code_desc_pairs = [
    ('CRASH CLASSIFICATION CODE', 'CRASH CLASSIFICATION DESCRIPTION'),
    ('MANNER OF IMPACT CODE', 'MANNER OF IMPACT DESCRIPTION'),
    ('PRIMARY CONTRIBUTING CIRCUMSTANCE CODE', 'PRIMARY CONTRIBUTING CIRCUMSTANCE DESCRIPTION')]

for code_col, desc_col in code_desc_pairs:
    mapping = crash_df[[code_col, desc_col]].dropna().drop_duplicates()
    code_to_desc = dict(zip(mapping[code_col], mapping[desc_col]))
    desc_to_code = dict(zip(mapping[desc_col], mapping[code_col]))
    crash_df.loc[crash_df[desc_col].isna() & crash_df[code_col].notna(), desc_col] = crash_df.loc[crash_df[desc_col].isna() & crash_df[code_col].notna(), code_col].map(code_to_desc)
    crash_df.loc[crash_df[code_col].isna() & crash_df[desc_col].notna(), code_col] = crash_df.loc[crash_df[code_col].isna() & crash_df[desc_col].notna(), desc_col].map(desc_to_code)
```

All discrepancies have been resolved, now proceeding to dealing with missing values

```python
crash_df.isna().sum()
```

```python
crash_df.isna().sum() / len(crash_df)
```

```python
percent_missing = (crash_df.isna().sum() / len(crash_df))[(crash_df.isna().sum() / len(crash_df)) > 0]
percent_missing
```

```python
percent_missing[percent_missing < 0.01]
```

```python
percent_missing[percent_missing.between(0.01, 0.05)]
```

```python
percent_missing[percent_missing > 0.05]
```

```python
row_drop = list(percent_missing[percent_missing < 0.01].index)
row_drop
```

```python
row_fill = list(percent_missing[percent_missing.between(0.01, 0.05)].index)
row_fill
```

```python
col_drop = list(percent_missing[percent_missing > 0.05].index)
col_drop
```

```python
crash_df = crash_df.drop(col_drop, axis = 1)
crash_df
```

```python
crash_df = crash_df.dropna(subset = row_drop)
crash_df
```

```python
for col in row_fill:
    crash_df[col] = crash_df[col].fillna(crash_df[col].mode(dropna = True)[0])
crash_df
```
